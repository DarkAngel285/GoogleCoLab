2) У нас есть словарь с номерами телефонов друзей и другой — с городами, где они живут, что не очень удобно. Хочется иметь один словарь, где ключом служит имя друга, а в значении лежит список из города и номера телефона. Напиши программу, которая объединит словари friend_base и phones в один и выведет его на экран.

```python
friend_base = {
'Киану Ривз': 'Тюмень', 'Джим Керри': 'Калуга', 'Сэмюэл Л. Джексон': 'Санкт-Петербург',
'Том Круз': 'Томск', 'Брэд Питт': 'Санкт-Петербург'
}
phones = {
'Киану Ривз': '+79033923029', 'Джим Керри': '+78125849204',
'Сэмюэл Л. Джексон': '+79053049385', 'Том Круз': '+79265748370'
}

new_base = {}

for name_friend, city in friend_base.items():
  for name_phones, phone in phones.items():
    if name_friend == name_phones:
      new_base[name_friend] = [city,phone]

print(new_base)
```

В этом уроке нам предстоит выступить в роли аналитика данных школы чародейства и волшебства “Магистериус”! Все упражнения мы будем выполнять на примере файла [StudentsPerformance.csv](https://https://disk.yandex.ru/d/7hQvwKr9KNp1Nw), в котором приводятся данные о результатах выпускных экзаменов.

```python
# считаем мальчиков и девочек
f = open('StudentsPerformance.csv')
males = 0
females = 0
for line in f:
  info = line.split(',') # каждую строку преобразуем в список
  gender = info[0] # выбираем первый элемент из строки
  if gender =='male':
    males = males + 1
  if gender =='female':
    females = females + 1
print("Мальчиков - ", males,"девочек - ", females)

f.close()
```

```python
# посчитаем оценки по зельеварению
potions = []

f = open('StudentsPerformance.csv')

for line in f:
    info = line.split(',')
    if info[0] == 'gender': # пропускаем заголовок таблицы
        continue
    else:
        mark = int(info[6]) # индекс оценки по зельеварению, переводим в числовой тип
        potions.append(mark) # добавляем оценку в список potions
potions_avg = sum(potions)/len(potions) # посчитаем средний балл
# посчитаем долю студентов, средний балл которых выше среднего
above_avg = 0
for mark in potions:
    if mark > potions_avg:
        above_avg += 1
print(above_avg / 1000)
```

```python
# превращаем весь файл в систему вложенных списков, каждый вложенный список будет содержать 1 строку данных и состоять из текста и чисел
exams = [] # создадим пустой список, который хранит все данные

f = open('StudentsPerformance.csv')
for line in f:
    info = line.split(',')
    if info[0] == 'gender': # пропускаем заголовок таблицы
        continue
    else:
      new_line = [] # храним значение каждой строки
      # разобъем элементы на два вида: текстовые и числовые (последний элемент не входит)
      for i in range(0,5):
        new_line.append(info[i])
      for i in range(5,8):
        new_line.append(int(info[i]))
      exams.append(new_line)
exams
```

В некоторых случаях при работе с очень большими текстовыми файлами возникает необходимость обобщения и группировки данных по определённым параметрам. Для решения такого типа задач, в случае если файл действительно большой и не помещается целиком в оперативную память, можно создать словарь, в котором будут храниться уже обработанные и сгруппированные данные. Такой словарь легко поместится в памяти компьютера, а извлечение из него нужных данных по ключам не составит большого труда.

```python
# преобразование файла в словарь
# словарь, содержащий информацию, какое кол-во студентов учится на разных факультетах
students = {} # создание пустого словаря

f = open('StudentsPerformance.csv')

for line in f:
    info = line.split(',')
    if info[0] == 'gender':
        continue
    else:
        faculty = info[1]
        # заполнение пустого словаря, проверяем ключ словаря
        if faculty in students:
            students[faculty] += 1
        else:
            students[faculty] = 1
students
```

В результате работы программа должна вывести на экран число лет, которое потребуется для накопления требуемой суммы. Ответ должен быть выведен на экран в виде целого числа без какого-либо дополнительного поясняющего текста.
